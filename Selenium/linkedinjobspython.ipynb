```python

# Get the full path to the Firefox driver executable
firefox_options = webdriver.FirefoxOptions()
# Create a Firefox WebDriver instance
driver = webdriver.Firefox(options=firefox_options)


```

```python
driver.implicitly_wait(10)
driver.get(url1)
```

```python
#Find number of job listings

y=driver.find_element(By.CLASS_NAME, 'results-context-header__job-count').text
# Given string
input_string = y

# Remove non-numeric characters
numeric_string = ''.join(char for char in input_string if char.isdigit())

# Convert the remaining string to an integer
result_integer = int(numeric_string)
y=result_integer
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


i = 2
while i <= int((n + 200) / 50) + 1:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    i += 1

    try:
        # Use WebDriverWait to wait for the "See more jobs" button to be clickable
        send = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='See more jobs']"))
        )
        send.click()
        time.sleep(3)

    except Exception as e:
        print(f"Error clicking the 'See more jobs' button: {str(e)}")
        break  # Break out of the loop in case of an error or if the button is not found
```

```python
#Create empty lists for company name and job title

companyname= []
titlename= []
# try
links=[]
```

```python
try:
    for i in range(n):
        companies = driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')
        
        # Check if the index is within the range of available elements
        if i < len(companies):
            company = companies[i].text
            companyname.append(company)
        else:
            print(f"No company found for index {i}")

except IndexError:
    print("IndexError: Unable to retrieve company names")
```

```python
try:
    for i in range(n):
        titles = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')
        
        # Check if the index is within the range of available elements
        if i < len(titles):
            title = titles[i].text
            titlename.append(title)
        else:
            print(f"No title found for index {i}")

except IndexError:
    print("IndexError: Unable to retrieve title names")
```

```python
#Create dataframe for company name and title

companyfinal=pd.DataFrame(companyname,columns=["Company"])
titlefinal=pd.DataFrame(titlename,columns=["Title"])
Joblink=pd.DataFrame(hrefList,columns=["Joblink"])


```
